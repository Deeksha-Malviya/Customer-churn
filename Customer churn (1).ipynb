```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)      
```

```python
df.head(20)
```

```python
plt.figure(figsize = (5,4))
axis = sns.countplot(x = 'Churn', data = df)
axis.bar_label(axis.containers[0])
plt.title("Count of customers by Churn")
plt.show()
```

```python
plt.figure(figsize = (3,3))
gb = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"],labels = gb.index, autopct = "%1.2f%%")
plt.title("% of Churned Customers", fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (5,4))
axis = sns.countplot(x = "gender", data = df, hue = "Churn")
axis.bar_label(axis.containers[0])
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (5,4))
axis = sns.countplot(x = "SeniorCitizen", data = df, hue = "Churn")
axis.bar_label(axis.containers[0])
plt.title("Churn by SeniorCitizen")
plt.show()
```

```python

# Assuming df is your DataFrame
plt.figure(figsize=(5, 4))
cross_tab = pd.crosstab(df["SeniorCitizen"], df["Churn"], normalize="index")

# Plot stacked bar chart
ax = cross_tab.plot(kind="bar", stacked=True, colormap="viridis", figsize=(5, 4))

plt.xlabel("Senior Citizen")
plt.ylabel("Proportion")
plt.title("Churn by SeniorCitizen (Stacked)")
plt.legend(title="Churn")
for bar in ax.containers:
    ax.bar_label(bar, fmt="%.2f", label_type="center", color="white", fontsize=10)
plt.show()
```

```python
# List of categorical columns
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create subplots
fig, axes = plt.subplots(3, 3, figsize=(15, 12))  # 3x3 grid
fig.suptitle("Count Plots for Different Services", fontsize=16)

# Flatten the axes array
axes = axes.flatten()

# Generate count plots for each column
for i, col in enumerate(columns):
    sns.countplot(x=df[col], ax=axes[i], palette="viridis", hue = df["Churn"])
    axes[i].set_title(col)
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to fit the title
plt.show()
```

```python
plt.figure(figsize = (10,4))
axis = sns.countplot(x = 'PaymentMethod', data = df, hue = "Churn")
axis.bar_label(axis.containers[0])
axis.bar_label(axis.containers[1])
plt.title("Count of customers by  PaymentMethod")
plt.show()
```
